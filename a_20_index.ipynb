```python
import numpy as np
import pandas as pd
```

```python
def a20_index(y_train,y_pred):
  total = 0
  acc = 0
  for i in range(len(y_train.values)):

    temp = np.abs((y_pred[i]-y_train.values[i])/y_train.values[i])
    if temp<=0.2:
      total += 1
      acc += 1
    else:
      total += 1

  a20 = acc/total
  return a20
```

```python
df = pd.read_excel("/content/drive/MyDrive/Projects/26. Iron Slug/Results.xlsx")
df.head()
```

```python
cols = df.columns
cols
```

```python
for i in range(1,len(cols)):
  a20 = a20_index(df[cols[0]],df[cols[i]])
  print(cols[i],a20)
```
